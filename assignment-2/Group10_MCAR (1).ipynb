#Little's MCAR test

```python
#dropping columns with more than half of values missing
rate_of_null = df_numeric.isnull().mean()
print(rate_of_null)
df_filtered = df_numeric.loc[:, rate_of_null < 0.5]
df_filtered
```

```python
#This leads to the dropping of months since last delinquet column
```

```python
#Check if any numeric data is left to process else raise error
if df_filtered.empty:
    raise Exception("No numeric data left to process for Little's test which is less than 50%")
```

```python
from scipy import stats

#function for a simplified MCAR test using chi-squares
def simple_chi_square_test(df):
  #convert to numpy
  data = df.to_numpy()

  missing_values = np.isnan(data)

  #return if no value to operate
  if not np.any(missing_values):
    print("No missing value found to operate, hence MCAR by definition")
    return

  # Calculate missing data pattern
  missing_patterns = np.sum(missing_values, axis=1)

  # Calculate statistics for determining missing data pattern
  unique_patterns, counts = np.unique(missing_patterns, return_counts=True)
  pattern_probabilities = counts / len(missing_patterns)

  # Perform chi-square test to check the pattern of the distribution
  chi2_stat, p_value = stats.chisquare(pattern_probabilities)

  print("~~~MCAR Test~~~")
  print(f"Chi-square statistic: {chi2_stat:.4f}")
  print(f"P-value: {p_value:.4f}")

  if p_value > 0.05:
      print("The null hypothesis cannot be rejected. The data is MCAR.")
  else:
      print("The null hypothesis is rejected. The data is not MCAR.")
```

```python
# Using the test on the filtered deteset
simple_chi_square_test(df_filtered)
```

#**Imputation using mean, median and mode**

```python
# Select only numeric columns for imputation
numeric_df = df.select_dtypes(include=np.number)
numeric_df
```

```python
# Check for missing values
missing_values = numeric_df.isnull().sum()
print(missing_values)
```

```python
#performing mean imputation
mean_imputed_data = numeric_df.fillna(numeric_df.mean())
print(mean_imputed_data.head())
print(mean_imputed_data.isnull().sum())
```

```python
#performing median imputation
median_imputed_data = numeric_df.fillna(numeric_df.median())
print(median_imputed_data.head())
print(median_imputed_data.isnull().sum())
```

```python
#performing mode imputation
mode_imputed_data = numeric_df.fillna(numeric_df.mode().iloc[0])
print(mode_imputed_data.head())
print(mode_imputed_data.isnull().sum())
```

```python
from fancyimpute import KNN, IterativeImputer
```

```python
#perform KNN imputation from fancyimpute
numeric_columns = numeric_df.select_dtypes(include=['number']).columns

knn_imputer = KNN(k=5)  # Setting the number of neighbors as needed

#Here we are only taking the first 1000 rows for KNN imputation as it is very heavy
#for running on such a large data set
df[numeric_columns][:1000]  = knn_imputer.fit_transform(df[numeric_columns][:1000])
```

```python
# Check the results
print(df[numeric_columns][:1000])
```

KNN does not work for categorical ddata so we have to encode it first

```python
# Encode categorical columns
df_encoded = pd.get_dummies(df.iloc[0:1000], drop_first=True)

# Perform imputation on the encoded DataFrame
df_knn_imputed = pd.DataFrame(KNN(k=5).fit_transform(df_encoded), columns=df_encoded.columns)
```

```python
df_knn_imputed.isnull().sum()
```

Now, performing iterative imputation (missing forest)

```python
# Perform Missing Forest imputation (Iterative Imputer)
# Select numerical columns for imputation
numeric_columns = df.select_dtypes(include=['number']).columns

# Performing missing forest imputation
missing_forest_imputer = IterativeImputer()
df[numeric_columns] = missing_forest_imputer.fit_transform(df[numeric_columns])

df.head()
```

```python
df.isnull().sum()
```

Similar to KNN, missing forest does not work for categorical data
